```python
import pandas as pd
import sys
import os
from datetime import date, datetime, timedelta
from dateutil.relativedelta import relativedelta
os.chdir('..')
sys.path.append('./economic_outlook/working_files')
from working_files.functions import *
from openbb import obb
obb.user.preferences.output_type = 'dataframe'

today = date.today()
last_five_years = today - relativedelta(years = 8)
```

```python
stocks = ['NVDA', 'AAPL', 'GOOGL', 'amd', 'pltr', 'fig']
```

```python
def data(list_stocks):
    # control flow, bettering the function
    list_stocks = [stock.upper() for stock in list_stocks if stock.lower()]
    # we save list of names in an empty list
    names_stocks = []
    # empty dataframe to init the data collection
    dataframe = pd.DataFrame()
    for stock in list_stocks:
        # reuse dataframe since we are recycling the dataframe, in a sense.
        dataframe = pd.concat([dataframe, obb.equity.price.historical(stock, provider = 'yfinance')['close']],axis=1)
        names_stocks.append(stock)
    dataframe.columns = names_stocks
    
    return dataframe
```

```python
data = data(stocks)
data
```
