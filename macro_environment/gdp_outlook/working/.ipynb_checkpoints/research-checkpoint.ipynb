```python
def finding_country(country_code):
    country = str(country_code).upper()
    data = df[df['code'].str.contains(country, na= False)]
    
    data.columns = [int(col) if col.isdigit() else col for col in df.columns]
    
    data = data.set_index('year') 
    return data
 
usa = (finding_country('USA'))

afg = (finding_country('afg'))
```

```python
def relative_years_data(country,begin_year, end_year = None):
    if end_year is not None:     
        df = country[(country.index > begin_year) & (country.index < end_year)]
    else:
        df = country[country.index > begin_year]
    return df

usa = moving_average(relative_years_data(usa, 2000, 2020),5)
afg = moving_average(relative_years_data(afg,2000,2020), 5)

usa
afg

usa
```

```python
def plotting_gdp(df):
    # Plot the two columns
    df[['gdp_per_capita', 'moving_average_across_5_years']].plot(figsize=(14, 6))

    # calculating the necessary data_points
    avg = df['gdp_per_capita'].mean()
    median = df['gdp_per_capita'].median()
    mean_idx = (df['gdp_per_capita'] - avg).abs().idxmin()
    median_idx = (df['gdp_per_capita'] - median).abs().idxmin()

    # getting the true index positions
    mean_y = df.loc[mean_idx, 'gdp_per_capita']
    median_y = df.loc[median_idx, 'gdp_per_capita']

    # Annotate mean
    plt.annotate(f'Mean: {mean_y:.2f}',
                 xy=(mean_idx, mean_y),
                 xytext=(mean_idx, mean_y + 1000),
                 arrowprops=dict(facecolor='green', shrink=0.05),
                 fontsize=10, color='green')

    # Annotate median
    plt.annotate(f'Median: {median_y:.2f}',
                 xy=(median_idx, median_y),
                 xytext=(median_idx, median_y + 1000),
                 arrowprops=dict(facecolor='blue', shrink=0.05),
                 fontsize=10, color='blue')

    plt.xlabel('Years')
    plt.ylabel('Gross Domestic Product Per Capita')
    plt.grid(True)
    plt.title('GDP against moving average')
    plt.show()
```

```python
plotting_gdp(usa)
plotting_gdp(afg)
```
